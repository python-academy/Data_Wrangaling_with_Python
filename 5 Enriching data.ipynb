**The Python code below reads and appends the datasets in the Appending data animation, above.**

```python
# Import packages
import pandas as pd
import numpy as np

# Import and display the first dataset
appendLeft = pd.read_csv("appending data.left.csv")
appendLeft
```

```python
# Import and display the second dataset
appendRight = pd.read_csv("appending data.right.csv")
appendRight
```

```python
# Join the first and second datasets.
# Sort the result along join columns.
appendLeft.merge(appendRight, how='outer', sort=True)
```

```python
# Concatenating rows
pd.concat([appendLeft, appendRight], axis=0)
```

```python
# Use apply to calculate the sum of EducationYears
appendLeft[['EducationYears']].apply(np.sum)
```

```python
# Insert a new column
z = appendLeft['EducationYears']>10

appendLeft.insert(loc=4, column='MoreThan10Years', value=z)
appendLeft
```
